# Скачиваем данные

```python
X = pd.read_csv('./market_data_train.csv')
```

```python
X
```

```python
X['symbol'].unique()
```

```python
y3 = pd.DataFrame()
y3['close_pct'] = X['close'].pct_change(3)

data3 = pd.concat([X, y3], axis = 1)
data3.dtypes
#y12 = X['close'].pct_change(12)
#y24 = X['close'].pct_change(24)
```

```python
data3 = data3.dropna()
y3 = data3['close_pct']
```

```python
data3
```

```python
X3 = data3
X3.drop('close_pct', axis = 1, inplace = True)
X3
```

```python
from sklearn.model_selection import train_test_split
X3_train, X3_val, y3_train, y3_val = train_test_split(X3, y3, test_size=0.10, random_state=42)
```

```python
X3_train
```

```python
y3_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
dtc = DecisionTreeClassifier(min_samples_split=4)
params = {'min_samples_split': [3,4,5,6,7,8,9,10]}
cv = StratifiedKFold(shuffle=True, n_splits=5)
```

```python
GS_dtc = GridSearchCV(estimator=dtc, param_grid=params, cv=cv, scoring='roc_auc')
```

```python
GS_dtc.fit(X=X3_train, y=y3_train)
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.model_selection import cross_validate,cross_val_score
```

```python
model_RF = RandomForestClassifier(random_state=1)
params = {'n_estimators': [100,500,1000, 1200,1500]}
cv = StratifiedKFold(shuffle=True, n_splits=5)
GS_RF = GridSearchCV(estimator=model_RF, param_grid=params, cv=cv, scoring='roc_auc')
GS_RF.fit(X3_train, y3_train)
model_RF = GS_RF.best_estimator_
```
