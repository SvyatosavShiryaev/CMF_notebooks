```python
X['close_pct'] = None
X['pnl_without_cumsum'] = None
X['pnl'] = None
X['close_pct'] = X['close'].pct_change(3)
# names = X.symbol.unique()
names = ['1INCH']
for name in names:
    print(name)
# X = X[X.symbol == '1INCH']
    X['close_pct'][X.symbol == name] = X['close'].pct_change(3)
    help_table = X[X.symbol == name]
    for i in range (1, len(help_table)):
        a = float(help_table['close'][help_table.index == i])
#        print(i)
        b = float(help_table['close'][help_table.index == i-1])
        help_table['pnl_without_cumsum'][help_table.index == i] = a - b
    X['pnl'][X.symbol == name] = help_table['pnl_without_cumsum'].cumsum()
X = X.drop('pnl_without_cumsum', axis = 1)
X
```

```python
X
```

```python
times = X.time.unique()
X['agregate_ind'] = None
#names = X.symbol.unique()
names = ['1INCH']
summa = 0
X['close_pct'] = X['close_pct'].fillna(0)
for t in times:
    for name in names:
#        print(0)
        l = list(X['close_pct'][(X.time == t) & (X.symbol == name)])
        if l == []:
            chislo = 0
        else:
            chislo = l[0]
        summa += chislo
#    print(summa)
    X['agregate_ind'][X.time == t] = summa/len(names)
        
```

```python
# Хотим посмотреть 12 фичей: для 4х типов цен и 3х видов параметров (всего 12)
price_type = ['open', 'high', 'low', 'close']
sma_hours = [3, 12, 24]
for p in price_type:
    for h in sma_hours:
        name = 'f_' + p + str(h)
        data = X[p]
        rolling_mean = data.rolling(window = h).mean()
        X[name] = X[p]/rolling_mean - 1
X
```

```python
pearsoncorr = X.corr(method='pearson')
pearsoncorr
```

```python
plt.rcParams['figure.figsize'] = (15.0, 10.0)
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
#             cmap='RdBu_r',
            annot=True,
            linewidth=1,
            fmt='.1g')
```
