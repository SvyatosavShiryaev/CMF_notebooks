```python
target_stat['time_diff'] = (target_stat['time1'] -  target_stat['time0'])/60
```

```python
target_stat['price_growth'] = target_stat['price1']/target_stat['price0']
```

```python
target_stat['given_target'].sum()
```

```python
target_stat['real_target'] = np.nan
```

```python
for i in range(0,len(target_stat['price_growth'])):
    if abs(target_stat['price_growth'][i]-1)>=0.05:
        target_stat['real_target'][i] = 1
    else:
        target_stat['real_target'][i] = 0
```

```python
target_stat['overlap'] = np.nan
for i in range(0,len(target_stat['price_growth'])):
    if target_stat['given_target'][i]==target_stat['real_target'][i]:
        target_stat['overlap'][i] = 1
    else:
        target_stat['overlap'][i] = 0
```
